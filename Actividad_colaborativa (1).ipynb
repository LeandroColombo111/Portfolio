```python
import pandas as pd
import matplotlib.pyplot as plt
# Leer el dataset
url='https://raw.githubusercontent.com/JJTorresDS/ds-data-sources/main/income_per_person_gdppercapita_ppp_inflation_adjusted.csv'
df= pd.read_csv(url,sep=',')
df
```

```python
# Filtrar los datos para obtener solo las filas correspondientes a Argentina
argentina_data = df[df['country'] == 'Argentina']

# Obtener el valor del GDP de Argentina en el año 2015
gdp_argentina1 = argentina_data['2015'].values[0]
gdp_argentina2 = argentina_data['2016'].values[0]
gdp_argentina3 = argentina_data['2020'].values[0]

print(f"El GDP de Argentina en 2015 fue: {gdp_argentina1}")
print(f"El GDP de Argentina en 2016 fue: {gdp_argentina2}")
print(f"El GDP de Argentina en 2020 fue: {gdp_argentina3}")
```

```python
# Función para convertir valores como '23k' a 23000
def convert_to_numeric(value):
    if isinstance(value, str) and 'k' in value:
        return float(value.replace('k', '')) * 1000
    elif isinstance(value, str):
        return float(value)
    else:
        return value

# Aplicar la función de conversión a las columnas de los años 2015 a 2021
for col in df.loc[:, '2015':'2021']:
    df[col] = df[col].apply(convert_to_numeric)

# Filtrar los datos para obtener solo las filas correspondientes a Argentina
argentina_data = df[df['country'] == 'Argentina']

# Seleccionar las columnas de los años 2015 a 2021 y calcular el promedio
argentina_gdp_values = argentina_data.loc[:, '2015':'2021']
argentina_gdp_mean = argentina_gdp_values.mean().mean()

print(f"El promedio del GDP de Argentina de 2015 a 2021 es: {argentina_gdp_mean:.2f}")
```

```python
# Función para convertir valores como '23k' a 23000
def convert_to_numeric(value):
    if isinstance(value, str) and 'k' in value:
        return float(value.replace('k', '')) * 1000
    elif isinstance(value, str):
        return float(value)
    else:
        return value

# Aplicar la función de conversión a las columnas de los años 2015 a 2021
for col in df.loc[:, '2015':'2021']:
    df[col] = df[col].apply(convert_to_numeric)

# Filtrar los datos para obtener solo los países de Sudamérica
south_america_countries = ['Argentina', 'Brazil', 'Colombia', 'Chile', 'Peru', 'Ecuador', 'Bolivia', 'Paraguay', 'Uruguay', 'Venezuela']
south_america_data = df[df['country'].isin(south_america_countries)]

# Filtrar los datos para obtener solo los países de América del Norte
north_america_countries = ['United States', 'Canada', 'Mexico']
north_america_data = df[df['country'].isin(north_america_countries)]

# Calcular el promedio del GDP para Sudamérica
sa_gdp_values = south_america_data.loc[:, '2015':'2021']
sa_gdp_mean = sa_gdp_values.mean().mean()

# Calcular el promedio del GDP para América del Norte
na_gdp_values = north_america_data.loc[:, '2015':'2021']
na_gdp_mean = na_gdp_values.mean().mean()

print(f"El promedio del GDP de Sudamérica de 2015 a 2021 es: {sa_gdp_mean:.2f}")
print(f"El promedio del GDP de América del Norte de 2015 a 2021 es: {na_gdp_mean:.2f}")
```
